# Enso vs Global SST at Equator

```python
#experimentSet = ['LME.002']
experimentSet = _model.cesmRcp85
```

```python

fig = plt.figure(figsize=(24,12))
plt.subplot(1,2,1)

for experiment in experimentSet:

    ds = xarray.load_dataset('results/cesmMonthlyIndeces/' + experiment + '.nc') 
    
    x=ds.nino34NoDetrend
    y=ds.backgroundSstNoDetrend

    plt.scatter(x,y, label=experiment)
    
plt.legend()
plt.title('No Detrend')
plt.xlabel('Nino34 No Detrend')
plt.ylabel('Background SST')                  
    
plt.subplot(1,2,2)

for experiment in experimentSet:

    ds = xarray.load_dataset('results/cesmMonthlyIndeces/' + experiment + '.nc') 
    
    x=ds.nino34
    y=ds.backgroundSstNoDetrend

    plt.scatter(x,y, label=experiment)
    
plt.legend()
                  
plt.title('With Detrend')

plt.xlabel('Nino34')
plt.ylabel('Background SST')  
```

```python
fig = plt.figure(figsize=(24,12))
plt.subplot(1,2,1)

for experiment in experimentSet:

    ds = xarray.load_dataset('results/cesmMonthlyIndeces/' + experiment + '.nc')
    
    x=ds.nino34
    y=ds.nino34NoDetrend

    plt.scatter(x,y, label=experiment)
    
plt.legend()

plt.title('Nino 34')
plt.xlabel('Detrend')
plt.ylabel('No Detrend')
```
